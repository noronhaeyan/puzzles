```python
class Solution:
    def totalNQueens(self, n: int) -> int:
        
        domain = [[0]*n for _ in range(n)]
        
        solution = [[0]*n for _ in range(n)]
        
        def print_domain(domain):
            n = len(domain)
            for i in range(n):
                print(domain[i])
        
        def in_domain(row, col, mode = ''):
            
            #debug
            if mode == 'debug':
                print("Checking domain for row: %s and col: %s" %(row, col))
                print_domain(domain)
            
            if domain[row][col] > 0:
                return True
            else:
                return False
        
        def place_queen(row, col, step = 1, mode = ''):
            curr_row = row
            curr_col = col
            
            if mode == 'debug':
                if step == 1:
                    print("Placing queen in row: %s and col: %s" %(row, col))
                else:
                    print("Removing queen in row: %s and col: %s" %(row, col))
                print("Before placement:")
                print_domain(domain)
            
            while row < n and col < n:
                domain[row][col] += step
                row += 1
                col += 1
            
            row = curr_row
            col = curr_col
            
            while row >= 0 and col >= 0:
                domain[row][col] += step
                row -= 1
                col -= 1
                
            row = curr_row
            col = curr_col
            
            while row < n and col >= 0:
                domain[row][col] += step
                row += 1
                col -= 1
                
            row = curr_row
            col = curr_col
            
            while row >= 0 and col < n:
                domain[row][col] += step
                row -= 1
                col += 1  
                
            for i in range(n):
                domain[i][curr_col] += step
                domain[curr_row][i] += step
                
#             print("during placement:", domain)
                
            domain[curr_row][curr_col] -= 5*step
            
            #debug
            if mode == 'debug':
                print("After placement:")
                print_domain(domain)
            
        def remove_queen(row, col, mode = ''):
            place_queen(row, col, step = -1, mode = mode)
               
        
        def backtrack(row = 0, count = 0, mode = ''):
            for col in range(n):
                if in_domain(row, col, mode = mode):
                    continue
                place_queen(row, col, mode = mode)
                solution[row][col] = 1
                if row == n - 1:
                    count += 1
                    
                    if mode == 'debug':
                        print("Solution found: ", solution)
                        
                    remove_queen(row, col)
                    break
                else:
                    count = backtrack(row+1, count, mode = mode)
                    remove_queen(row, col, mode = mode)
                    solution[row][col] = 0
                    
            return count
                    
                
        
        return backtrack(mode = '')
```
